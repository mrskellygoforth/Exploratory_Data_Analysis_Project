# Importing Libraries
The following cell imports the necessary libraries for exploring and analyzing the data.

### numpy
This library includes mathematical functions that are more complex than just basic operations that are inherent to python, including trigonometric, exponential, and radical functions.  This library is also used for creating arrays and matrices, which allows for the implementation of linear algebra.

### pandas
This library is used for creating and manipulating series and dataframes, which are similar to arrays and matrices, except they allow for each column in the data set to have a different data type.  This library also includes functionalities similar to SQL, such as merge and groupby.

### matplotlib.pyplot
The pyplot is a module within the matplotlib library that is used for creating univariate and multivariate plots, such as histograms, bar plots, box plots, and scatter plots.  The addition of "%matplotlib inline" tells python to not only create the plot, but also display it.

### seaborn
This library is built upon the matplotlib library, but allows for the creation of more visually appealing plots.  The addition of "sns.set(color_codes = True)" applies seaborn's default color palette.

### warnings
Filtering the warnings to 'ignore' will prevent warnings about potential future issues from appearing when cells are executed.

```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set(color_codes = True)

import warnings
warnings.filterwarnings('ignore')
```

## MILES, USAGE, AND FITNESS LEVEL

#### Miles According to Usage
• As one would expect, there is a strong positive relationship between the number of times per week the customer expects to use the treadmill and the number of miles they expect to run.  Generally speaking, the more times per week a customer uses the treadmill, the more miles they will run.

• The range of the number of miles run in each 'Usage' category is between 100-150 miles.  One noteable exception to this is the outlier in the 'Usage' = 4 category, where one customer reports that they expect to run over 350 miles, despite only using the treadmill 4 times a week.  The next highest number of miles in this 'Usage' category is only about 175 miles.  We may want to ignore this outlier, or investigate it further.

#### Miles According to Fitness Level
• As one would expect, there is a strong positive relationship between a customer's self-reported fitness level and the number of miles they expect to run.  Generally speaking, the higher a customer's fitness level, the more miles they will run.

• The range of the number of miles run in each 'Fitness' category increases as the fitness level increases.  For example, those who reported a fitness level of 2 expect to run between approximately 30 miles and approximately 80 miles (a range of 50 miles).  However, those who reported a fitness level of 5 expect to run between approximately 75 and approximately 360 miles ( a range of 285 miles).  This may indicate that profiling customers with a lower fitness level may be more accurate because there exists less spread in those groups.

#### Usage According to Fitness Level
• As one would expect, the higher a customer's fitness level, the more they will use the treadmill per week.  For example, those with a fitness level of 1 or 2 will only use the treadmill 2 or 3 times per week.  However, those with a fitness level of 5 will use the treadmill 3-7 times per week.  

#### Products purchased based on Usage, Fitness Level, and Miles
• Customers with higher fitness levels, who plan on using the product more often, and who intend on running more miles tend to prefer the TM798.
• Customers with lower fitness levels, who plan on using the product less often, and who intend on running fewer miles tend to prefer the TM195 or the TM498.
